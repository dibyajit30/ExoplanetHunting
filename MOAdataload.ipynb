```
from tqdm import tqdm
```

```
gb18=[]
```

```
gb18=[]
pbar=tqdm(p.glob('**/*.phot'))
for f in pbar:
    pbar.set_description(f'processing {f}')
    g=pd.read_csv(f, skiprows=1, delim_whitespace=True, header=None, names=columns)[:-1]
    chip=str(f).split('/')[2]
    ab_mag=-2.5 * np.log(g['flux'] / (691.8 if chip=='2' else 1445)) + 20
    g['ab_mag']=ab_mag
    gb18.append([str(f), g])
```
